# Explore Data

This Jupyter Notebook is used to explore the given data sets.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import glob
import os
%matplotlib inline
```

## Helper Functions

```python
def show_images(img_paths, rows=5, columns=4, figsize=(64,64)):
    fig=plt.figure(figsize=figsize)
    for i in range(0, columns*rows):
        img_path = img_paths[i]
        img = mpimg.imread(img_path)
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(img)
    plt.show()
```

## GTI

```python
data_folder = os.path.join('data', 'vehicles')
gti_data = ['GTI_Far', 'GTI_Left', 'GTI_MiddleClose', 'GTI_Right']
gti_paths = [os.path.join(data_folder, gti) for gti in gti_data]

path_list = []
for gti_path in gti_paths:
    path = os.path.join(gti_path, '*.png')
    paths = glob.glob(path)
    print(gti_path)
    print('\t{} elements'.format(len(paths)))
    path_list += paths
    
print('\n')
print('total: {} elements'.format(len(path_list)))
```

```python
# Show first 20 of list to see if there are time series dependencies
show_images(path_list[:20])
```

There are multiple images of the same car with only minor variations. This might lead to problems later.
